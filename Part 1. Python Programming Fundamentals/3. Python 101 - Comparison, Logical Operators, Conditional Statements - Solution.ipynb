# 1. 比較演算子


````
# 比較演算子の出力は "True "または "False "である。
# 最初に equal '==' 比較演算子を取り上げよう
# 単純に質問をするだけだ。"stock_1_priceはstock_2_priceと等しいか？"
# "True "は条件を満たすことを意味する 
# "False "と出力された場合、条件が満たされていない（条件が真ではない）ことを意味する 


stock_1_price = 1000
stock_2_price = 1000
stock_1_price == stock_2_price
````

````
# 大なり小なりの演算子 '>='

stock_1_price = 5000
stock_2_price = 1000
stock_1_price >= stock_2_price
````

````
# ♪ '==' は比較演算子であることに注意してください 
# なお、'='は変数の代入に使われます（stock_1_priceに10を入れる）。
stock_1_price = 10
````

````
stock_1_price
````

````
# 比較演算子は文字列でも動作します
'Finance' == 'Finance'
````

````
# 両方の文字列が完全に一致する必要がある (小文字と大文字)
'Finance' == 'finance'
````

````
# 等しくない比較演算子 '!='

stock_1_price = 1000
stock_2_price = 1000
stock_1_price != stock_2_price
````

# 2.  論理演算子

````
# "and "論理演算子では、出力が "True "になるためには、両方の引数が "True "でなければなりません。
# (注意: 'and' 論理演算子が True の出力を生成する唯一のケースです)
# 1 and 1 = 1
True and True 
````

````
# 0 and 1 = 0
False and True 
````

````
# 0 and 0 = 0
False and False 
````

````
# "or "論理演算子では、引数の1つでも "True "であれば、出力は "True "となる 
# 1 or 1 = 1
True or True 
````

````
# 1 or 0 = 1
True or False 
````

````
# 0 or 1 = 1
False or True 
````

````
# 0 or 0 = 0 (これは'or'論理演算子がFalseの出力を生成する唯一のケースであることに注意)
False or False 
````

````
 # True and False = False 
(10>3) and (9==3)
````

````
# True or False = True
(9>3) or (12==10) 
````

# 3. 条件文

- 単純なif-else文をPythonで書くと以下のようになります。

```
if condition:
  statement #1
else:
  statement #2
```
- 条件が真であれば、インデントされた最初のステートメントを実行します。
- 条件が真でない場合は，インデントされたelse文を実行する。
- Pythonはインデント（空白）を使ってコードのセクションやスコープを示すことに注意してください。


````
if 10 > 9:
    print('If condition is True')
else:
    print('If condition is False')
````

````
if 5 == 10:
    print('If condition is True')
else:
    print('If condition is False')
````

````
if 5 == 7:
    print('First statement is true')
elif 2 == 4:
    print('Second statement is true')
else:
    print('Last statement is true')
````

````
if 5 < 7:
    print('First statement is true')
elif 2 > 4:
    print('Second statement is true')
else:
    print('Last statement is true')
````

````
# ユーザーからの入力を受け取り、それに応じてアクセスを許可または拒否しよう
name = input('Enter your username:')

if name == 'Ryan':
    print("Access granted")
else:
    print("Access Denied")
````

````
# コードをより強固なものにしよう（大文字小文字を区別しない）
name = input('Enter your username:')

if name == 'Ryan' or name =='RYAN' or name =='ryan':
    print("Access granted")
else:
    print("Access Denied")
````
