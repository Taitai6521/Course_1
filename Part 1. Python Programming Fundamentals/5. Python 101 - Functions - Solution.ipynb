# 3. ラムダ式 

```
# ラムダ関数は名前のない関数を作るために使われる 
# ラムダ関数は主にfilter()やmap()で使用されます[後述します]。
# ラムダ関数は任意の数の引数を受け取ることができますが、1つの式しか持つことができません。

```

```
# 入力を二乗する基本的な関数の作り方を見てみましょう。
def my_function(x):
    return x**2
```

```
# 関数の呼び出し
my_function(3)
```

```
# 同じ作業をLambda式で行うことができます。
# 関数名がないことに注意

y = lambda x:x**2
```

```
y(3)
```

```
# ラムダ式は複数の引数を取れることに注意 
# 実行可能な表現は一つしかない

def summation(x, y, z):
    return x+y+z
```

```
result = summation(1, 2, 3)
result
```

```
result = lambda    x, y, z : x+y+z   
```

```
result(1, 2, 3)
```

# 4. MAP 

```
# map()関数は、関数とリストを受け取ります。
# map()は、リスト全体に対して演算を行い、その結果を新しいリストで返します。
# 2つのリストaとbを定義する

a = [1, 4, 5, 6, 9]
b = [1, 7, 9, 12, 7]
```

```
# 2つの要素を加算する関数を定義しよう
def summation(a, b):
    return a + b
```

```
# map()を使ってリスト全体に関数を適用し、新しいリストを生成することができます。
c = list(map(summation, a, b))
c
```

```
# mapは次のようにLambdaと一緒に使うことができます。
# lambda関数は、任意の数の引数を取り、1つの式しか持てない無名関数です。

input_list = [1, 2, 3, 4]

out_list = list(map ( lambda x:x**2 , input_list))
out_list
```

# 5. フィルタ

```
# フィルタは、関数が "True "を返す要素のリストを作成するために使用されます。
prices = [105, 5055, 40, 356, 923, 1443, 222, 62]
```

```
# 偶数のみを返す
out = list(filter( lambda x: (x % 2 == 0) , prices  ))
out
```

```
# 100以上の価格を返す
prices = [105, 5055, 40, 356, 923, 1443, 222, 62]
out = list(  filter(lambda x: (x>=100) , prices) )
out
```

```
# 200から250の値を返す
out = list(  filter(lambda x: (x>=200 and x<=250) , prices) )
out
```

**ミニチャレンジ2：解答 
- ユーザーから2つの入力（ユニット数と価格）を受け取り、アカウントの合計残高を計算するコードを書いてください。


```
def total(units, price):
    amount = units * price
    return(amount)
```

```
num_shares = int(input("Enter the number of shares: "))
price = int(input("Enter the price per share: "))
account_balance = total(num_shares, price)

print('Total account balance = {}'.format(account_balance))
```

**ミニチャレンジ#3の解決策**。
- ミニチャレンジ#2（ユーザーから2つの入力を受けて合計を計算するコードを書く）をラムダ式を使って再現します**。


```
num_shares = int(input("Enter the number of shares: "))
price = int(input("Enter the price per share: "))

account_balance = lambda num_shares, price : num_shares * price
print('Total = {}'.format(account_balance(num_shares, price)))
```

**ミニチャレンジ #4 解決策:**
- **引数を取り、その3乗値を返す関数を書きます**。
- **-10から10までの整数のリストを定義する**。
- **リスト全体に関数を適用し、新しい出力リストを生成する**。


```
def cubic(x):
    return (x*x*x)

numbers = range(-10, 11)
numbers_cubed = list(map(lambda x: cubic(x), numbers))
print(list(numbers))
print(numbers_cubed)
```

**ミニチャレンジ#5の解答:**
- 範囲（上下限）を指定して、正の数と偶数のみのリストを返すコードを書いてください。


```
min = int(input("Enter a number (lower bound) *must be negative* : "))
max = int(input("Enter a number (Upper bound:) "))

numbers = range(min, max)
even_greater_than_zero = list(filter(lambda x: (x > 0 and x%2 == 0), numbers))
print(even_greater_than_zero)
```
