# 2. LIST

````
# リストとは、順序付けされた変更可能なコレクションです。
# リストは重複したメンバーを許可します。

my_list = [1, 2, 3]
my_list
````

````
# データ型の取得
type(my_list)
````

````
# データ型が混在するリスト (例えば、文字列と整数を1つのリストに入れることができます)
my_list = [2, "GOOG", 5]
my_list
````

````
# リストの中に別のリストを入れることができます(ネストされたリスト)
my_list = ["GOOG", [3, 6, 7]]
my_list
````

````
# インデックスを使ってリストの特定の要素にアクセスする
# リストの最初の要素のインデックスは0であることに注意してください（少し混乱しますが、すぐに慣れるでしょう）

my_list = [2, "GOOG", 5]
my_list[1]
````

````
# ネストしたリストのインデックス付け
my_list = ["GOOG", [3, 6, 7], ["APPL", 5, 6]]
my_list[2][0]
````

````
print(my_list[1][1])
````

````
print(my_list[2])
````

````
# 負のインデックス
my_list = [2, "GOOG", 5]
print(my_list[-2])
````

````
# リストスライス (リストから複数の要素を取得する) 
my_list = [2, 5, 7, "GOOG", "SP500", "AMZN", 15]

# 添字3から始まり、添字6の要素を含まない要素を得る  
print(my_list[3:6])
````

````
# 添字3から始まる要素を最後まで取得する
print(my_list[3:])
````

````
# 始まりから終わりまでの全ての要素
print(my_list[:])
````

````
# リストの長さ（リストに含まれる要素の数）を求める 
len(my_list)
````

# 3. ディクショナリー

````
# my_dict = {'key1':'value1', 'key2':'value2', 'key3':'value3'}
# 辞書は、キーと値のペアの集まりで構成される。各key-valueペアは、キーとそれに対応する値をマッピングします。
# キーは辞書内で一意であるが、値は一意でない場合がある。
# リストの要素はインデックスを使ってリスト内の位置でアクセスされるが、辞書の要素はキーでアクセスされる。
# キーと値のペアで辞書を定義する

my_dict = {'key1':'value1',
           'key2':'value2', 
           'key3':'value3'}
````

````
# 辞書の表示
my_dict
````

````
# データ型の確認
type(my_dict)
````

````
# 辞書の特定の要素に、特定のキーを使ってアクセスする (例: Key2)
my_dict['key2']
````

````
# すべての資産をリストアップした銀行の顧客のための辞書を定義する! 
client_assets = {"Real estate": 150000,
                 "stocks": 10000,
                 "savings": 20000}

print(client_assets)
````

````
# 辞書の要素にアクセス
client_assets["stocks"]
````

````
# 既存の辞書に新しい項目を追加する
client_assets["Crypto"] = 1500
print(client_assets)
````

````
# 辞書から項目を削除
del client_assets["savings"]
print(client_assets)
````

````
# 辞書全体を削除する
# 辞書全体が削除されているので、再度プリントアウトすることはできません。

del client_assets
print(client_assets)
````

# 5. タプル

````
# タプルは不変のPythonオブジェクトのシーケンスです。
# タプルはリストのようなシーケンスです。
# タプルとリストの違いは、タプルはリストと違って変更できないことと、タプルは括弧を使うのに対し、リストは角括弧を使うことです。
````

````
tuple_1 = ('GOOG', 'APPL', 10, 15, 1992);
tuple_2 = (450, 55, 977, 2100);
type(tuple_1)
````

````
tuple_1
````

````
# ♪タプルの要素にアクセスする
tuple_1[1]
````

````
# タプルのスライシング
tuple_1[1:3]
````

````
# タプルの値を変更してもうまくいかない! 
# 'tuple' オブジェクトはアイテムの割り当てをサポートしていない

tuple_1[1] = 0
````

````
# 既存のタプルから新しいタプルを作成する
tuple_3 = tuple_1 + tuple_2
tuple_3
````

# 6. セット 

````
# セットとは、順序付けされていないアイテムの集まりです。すべての要素は一意である（重複しない）。
# セットは、すべてのアイテム（要素）を中括弧{}で囲み、カンマで区切るか、組み込み関数set()を使って作成します。
# 整数のセット

my_set = {'GOOG', 'APPL', 'T'}
print(my_set)
````

````
# 重複していないセット
my_set = {'GOOG', 'APPL', 'T','TSLA','T','AAPL'}
print(my_set)
````
